```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42

color_map = ['#014f3f','#01665e', '#5da8a1', '#80cdc1', '#c7eae5']


#The function is modified from https://github.com/BigDataBiology/SemiBin_benchmark/blob/main/visualization/CAMI_I_comparison.py
def get_number_of_genomes_per_completeness(amber_path, return_pandas=False):
    genome_path = os.path.join(amber_path, 'genome')
    table = {}
    for root, dirs, files in os.walk(genome_path, topdown=False):
        for name in dirs:
            method_path = os.path.join(root, name)
            metric = pd.read_csv(os.path.join(method_path, 'metrics_per_bin.tsv'), sep='\t')
            metric = metric.query('`Purity (bp)` >= 0.95')
            com_90_pur_95 = metric.eval('`Completeness (bp)` > 0.9').sum()
            com_80_pur_95 = metric.eval('`Completeness (bp)` > 0.8').sum() - (
                com_90_pur_95)
            com_70_pur_95 = metric.eval('`Completeness (bp)` > 0.7').sum() - (
                    com_90_pur_95 + com_80_pur_95)
            com_60_pur_95 = metric.eval('`Completeness (bp)` > 0.6').sum() - (
                    com_90_pur_95 + com_80_pur_95 + com_70_pur_95)
            com_50_pur_95 = metric.eval('`Completeness (bp)` > 0.5').sum() - (
                    com_90_pur_95 + com_80_pur_95 + com_70_pur_95 + com_60_pur_95)
            table[method_path.split('/')[-1]]= [com_90_pur_95, com_80_pur_95, com_70_pur_95, com_60_pur_95, com_50_pur_95]
    if return_pandas:
        return pd.DataFrame(table, index=[90,80,70,60,50]).T
    return table
```

```python
#The function is modified from https://github.com/BigDataBiology/SemiBin_benchmark/blob/main/visualization/CAMI_I_comparison.py
#
def plot_bar(amber_path, add_legend=True,y_label = None,title = None, output = None):
    data = get_number_of_genomes_per_completeness(amber_path, return_pandas=True)

    if 'COMEBin_filter200000' in data.index:
            data = data.rename(index={'COMEBin_filter200000': 'COMEBin'})
    elif 'COMEBin(filter200000)' in data.index:
            data = data.rename(index={'COMEBin(filter200000)': 'COMEBin'})
    elif 'COMEBinfilterminszie200000' in data.index:
        data = data.rename(index={'COMEBinfilterminszie200000': 'COMEBin'})

    if 'Metadecoder' in data.index:
        data = data.rename(index={'Metadecoder': 'MetaDecoder'})

    subset = data.loc[[
    'COMEBin',
    'SemiBin2',
    'SemiBin1',
    'MetaDecoder',
    'CLMB',
    'VAMB_v413',
    'MetaBAT',
    'MaxBin',
    'CONCOCT']]

    subset.rename(index={
        'MaxBin': 'MaxBin2',
        'MetaBAT': 'MetaBAT2',
        'VAMB_v413': 'VAMB',
    }, inplace=True)

    subset = subset[[90,80,70,60,50]]
    print(subset)
    high_quality_list = subset[90].sort_values().values
    print('Improvement of best binner over second best: {:.2%}'.format(
        (high_quality_list[-1] - high_quality_list[-2]) / high_quality_list[-2]))
    ax = subset.plot(kind="barh", stacked=True, legend=False, color=color_map)

    if add_legend:
        ax.legend(['>90', '>80', '>70', '>60','>50'],
                  loc='upper right', fontsize=10, title='completeness')
    ax.set_xticks(ticks=y_label)
    ax.set_xticklabels(labels=y_label, fontsize=15, color='black')
    ax.set_yticklabels(labels=subset.index, fontsize=15, color='black')
    ax.set_xlabel('#Bins(< 5% contamination)', fontsize=15, color='black')
    ax.set_title(title, fontsize=15, alpha=1.0)

    plt.tight_layout()
    plt.savefig(output, dpi=300)
    plt.show()
    plt.close()
```

```python
def get_number_of_genomes_per_completeness_checkm2(checkm2_res_file, return_pandas=False):#
    metric = pd.read_csv(checkm2_res_file,index_col=0, sep='\t')
    table = {}
    for bin_method in list(metric.index):
        if bin_method != 'Method':
            com_90_pur_95 = int(metric['num_905'][bin_method])
            com_80_pur_95 = int(metric['num_805'][bin_method]) - com_90_pur_95
            com_70_pur_95 = int(metric['num_705'][bin_method]) - com_90_pur_95 - com_80_pur_95
            com_60_pur_95 = int(metric['num_605'][bin_method]) - com_90_pur_95 - com_80_pur_95 - com_70_pur_95
            com_50_pur_95 = int(metric['num_505'][bin_method]) - com_90_pur_95 - com_80_pur_95 - com_70_pur_95 - com_60_pur_95
            table[bin_method.split('/')[0]]= [com_90_pur_95, com_80_pur_95, com_70_pur_95, com_60_pur_95, com_50_pur_95]

    if return_pandas:
        return pd.DataFrame(table, index=[90,80,70,60,50]).T
    return table



def plot_bar_realdata(checkm2_res_file, add_legend=True,y_label = None,title = None, output = None):
    data = get_number_of_genomes_per_completeness_checkm2(checkm2_res_file, return_pandas=True)

    if 'concoct_bins_checkm2_output' in data.index:
        data = data.rename(index={'concoct_bins_checkm2_output': 'CONCOCT'})
    if 'maxbin2_bins_checkm2_output' in data.index:
        data = data.rename(index={'maxbin2_bins_checkm2_output': 'MaxBin2'})
    if 'metabat2_bins_checkm2_output' in data.index:
        data = data.rename(index={'metabat2_bins_checkm2_output': 'MetaBAT2'})
    if 'vamb_v413_m200000' in data.index:
        data = data.rename(index={'vamb_v413_m200000': 'VAMB'})
    if 'clmb_m200000' in data.index:
        data = data.rename(index={'clmb_m200000': 'CLMB'})
    if 'metadecoder' in data.index:
        data = data.rename(index={'metadecoder': 'MetaDecoder'})
    if 'semibin2_Easy_coassemly_mode' in data.index:
        data = data.rename(index={'semibin2_Easy_coassemly_mode': 'SemiBin2'})
    if 'semibin_Easy_coassemly_mode_v1.0.0_default' in data.index:
        data = data.rename(index={'semibin_Easy_coassemly_mode_v1.0.0_default': 'SemiBin1'})
    if 'COMEBin_nocovloss_tau0.15_nepoch200_earlystop_addvars_nedge75_vars_sqrt_NokmerMetric' in data.index:
        data = data.rename(index={'COMEBin_nocovloss_tau0.15_nepoch200_earlystop_addvars_nedge75_vars_sqrt_NokmerMetric': 'COMEBin'})

    subset = data.loc[[
        'COMEBin',
        'SemiBin2',
        'SemiBin1',
        'MetaDecoder',
        'CLMB',
        'VAMB',
        'MetaBAT2',
        'MaxBin2',
        'CONCOCT']]

    subset = subset[[90,80,70,60,50]]
    print(subset)
    high_quality_list = subset[90].sort_values().values
    print('Improvement of best binner over second best: {:.2%}'.format(
        (high_quality_list[-1] - high_quality_list[-2]) / high_quality_list[-2]))
    ax = subset.plot(kind="barh", stacked=True, legend=False, color=color_map)

    if add_legend:
        ax.legend(['>90', '>80', '>70', '>60','>50'],
                  loc='upper right', fontsize=10, title='completeness')
    ax.set_xticks(ticks=y_label)
    ax.set_xticklabels(labels=y_label, fontsize=15, color='black')
    ax.set_yticklabels(labels=subset.index, fontsize=15, color='black')
    ax.set_xlabel('#Bins(< 5% contamination)', fontsize=15, color='black')
    ax.set_title(title, fontsize=15, alpha=1.0)

    plt.tight_layout()
    plt.savefig(output, dpi=300)
    plt.show()
    plt.close()
```

```python
#The codes are modified from https://github.com/BigDataBiology/SemiBin_benchmark/blob/main/visualization/Real_datasets_comparison.py
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['svg.fonttype'] = 'none'

def get_taxa_sets(bac_path, arr_path = None):
    bac = pd.read_csv(bac_path,
                      '\t').values
    if arr_path is not None:
        arr = pd.read_csv(arr_path,
                          '\t').values
        bac = np.concatenate((bac,arr),axis=0)
    print(bac.shape)
    species_list = []
    genus_list = []
    family_list = []
    order_list = []
    Class_list = []
    phylum_list = []
    domain_list = []

    for temp in bac:
        #        print(temp)
        split = temp[1].split(';')
        #        print(len(split))
        if len(split) == 1:
            continue
        else:
            species = split[-1]
            genus = split[-2]
            family = split[-3]
            order = split[-4]
            Class = split[-5]
            phylum = split[-6]
            domain = split[-7]
            if species[0:3] != 's__' or genus[0:3] != 'g__' or family[0:3] != 'f__' or order[0:3] != 'o__' or Class[0:3] != 'c__' or phylum[0:3] != 'p__' or domain[0:3] != 'd__':
                print('error')

            if species != 's__':
                species_list.append(species)
            if genus != 'g__':
                genus_list.append(genus)
            if family != 'f__':
                family_list.append(family)
            if order != 'o__':
                order_list.append(order)
            if Class != 'c__':
                Class_list.append(Class)
            if phylum != 'p__':
                phylum_list.append(phylum)
            if domain != 'd__':
                domain_list.append(domain)

    return set(domain_list),set(phylum_list),set(Class_list),set(order_list), set(family_list), set(genus_list), set(species_list)
```

```python

# def plot_taxa_overlap_split_5005_merge():
COLORS = {
    'both': '#7570b3',
    'comebin': '#1b9e77',
    'metabat2': '#ec7014',
}

fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.flatten()

for i, dataset in enumerate(['watergroup2', 'watergroup3', 'MetaHIT_reads','MetaHIT_reads_coassembly_10sample']):
    #for dataset,ax in zip(['watergroup2', 'watergroup3', 'MetaHIT_reads'], axes.flat):
    if dataset == 'watergroup2' or dataset == 'watergroup3':
        comebin_taxa = get_taxa_sets('../results/real_data/gtdb_res/{0}/COMEBin/checkm2_quality_5005_gtdbtk_output/gtdbtk.bac120.summary.tsv'.format(dataset))

        Metabat2_taxa = get_taxa_sets('../results/real_data/gtdb_res/{0}/metabat2_bins/checkm2_quality_5005_gtdbtk_output/gtdbtk.bac120.summary.tsv'.format(dataset))
    if dataset == 'MetaHIT_reads' or dataset =='MetaHIT_reads_coassembly_10sample':
        comebin_taxa = get_taxa_sets('../results/real_data/gtdb_res/{0}/COMEBin/checkm2_quality_5005_gtdbtk_output/gtdbtk.bac120.summary.tsv'.format(dataset),
                                     '../results/real_data/gtdb_res/{0}/COMEBin/checkm2_quality_5005_gtdbtk_output/gtdbtk.ar53.summary.tsv'.format(dataset))

        Metabat2_taxa= get_taxa_sets('../results/real_data/gtdb_res/{0}/metabat2_bins/checkm2_quality_5005_gtdbtk_output/gtdbtk.bac120.summary.tsv'.format(dataset),
                                     '../results/real_data/gtdb_res/{0}/metabat2_bins/checkm2_quality_5005_gtdbtk_output/gtdbtk.ar53.summary.tsv'.format(dataset))

    assert len(comebin_taxa) == len(Metabat2_taxa)
    subset = pd.DataFrame([(len(sb & mb2), len(sb - mb2), len(mb2 - sb)) for sb,mb2 in zip(comebin_taxa, Metabat2_taxa)],
                          index=['Domain', 'Phylum', 'Class', 'Order', 'Family', 'Genus', 'Species'],
                          columns=['Both', 'COMEBin only', 'MetaBAT2 only'])
    subset = subset.iloc[::-1]

    ax = axes[i]

    ax.bar(np.arange(len(subset))-.15, subset.Both + subset['MetaBAT2 only'], width=.3, color=COLORS['metabat2'], label='MetaBAT2 only')
    ax.bar(np.arange(len(subset))+.15, subset.Both + subset['COMEBin only'], width=.3, color=COLORS['comebin'], label='COMEBin only')
    ax.bar(np.arange(len(subset)), subset.Both, width=.6, color=COLORS['both'], label='Both')
    title = {
        'watergroup2': 'Water Group2',
        'watergroup3': 'Water Group3',
        'MetaHIT_reads': 'MetaHIT',
        'MetaHIT_reads_coassembly_10sample':'MetaHIT (10-sample)'
    }[dataset]
    ax.set_title(title, fontsize=10, alpha=1.0, color='black')
    ax.set_ylabel('Number of taxa', fontsize=10, color='black')

    if i==0:
        ax.legend(loc='upper right', fontsize=8)

    ax.set_xticks(np.arange(len(subset)))
    ax.set_xticklabels(labels=subset.index, rotation=30, minor=False, fontsize=10, color='black')

plt.tight_layout()
plt.savefig('./figures/combined_plots_5005_overlap_res.pdf', dpi=300, bbox_inches='tight')
plt.show()
plt.close()

```

```python
import pandas as pd
from collections import Counter
import numpy as np
import os
import seaborn as sns
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

```python
labelfile = '../results/features/evaluation_285047.txt'
label_df = pd.read_csv(labelfile,sep='\t')


BINID_list=label_df['BINID']

c = Counter(BINID_list)

top_10_bin = c.most_common(10)

top_10_bin = [item[0] for item in top_10_bin]
# ['OTU_97.37995.0',
#  'OTU_97.908.0',
#  'OTU_97.343.0',
#  'OTU_97.1426.0',
#  'OTU_97.156.0',
#  'OTU_97.38219.0',
#  'OTU_97.391.0',
#  'OTU_97.5863.0',
#  'OTU_97.41724.0',
#  'OTU_97.6209.0']


label_df_top10bin=label_df[label_df['BINID'].isin(top_10_bin)]


seq_to_binid=dict(zip(label_df_top10bin["@@SEQUENCEID"],label_df_top10bin['BINID']))

contigs_top10bin=label_df_top10bin['@@SEQUENCEID']

contigs_top10bin_list=list(contigs_top10bin)
```

```python
#（raw kmer; k=4)
```

```python
def gen_kmer(com_file):
    namelist = pd.read_csv(com_file, sep=',', usecols=range(1)).values[:, 0]
    compositHeader = pd.read_csv(com_file, sep=',', nrows=1)
    compositMat = pd.read_csv(com_file, sep=',', usecols=range(1, compositHeader.shape[1])).values

    compositMat = compositMat + 1
    compositMat = compositMat / compositMat.sum(axis=1)[:, None]

    return compositMat, namelist

kmer_4_file='../results/features/kmer_4_f0.csv'
data_4mer, namelist_4mer = gen_kmer(kmer_4_file)
kmer_4_df=pd.DataFrame(data_4mer,index=namelist_4mer)


kmer_4_top10bin_df=kmer_4_df[kmer_4_df.index.isin(contigs_top10bin_list)]



kmer_4_top10bin_label=[]
for key in list(kmer_4_top10bin_df.index):
    kmer_4_top10bin_label.append(seq_to_binid[key])


# Visualize embeddings using tSNE
comb_emb = kmer_4_top10bin_df.values
comb_lab = np.array(kmer_4_top10bin_label)




comb_tsne = TSNE().fit_transform(comb_emb)

fig, ax = plt.subplots()


sns.scatterplot(
    x=comb_tsne[:, 0],
    y=comb_tsne[:, 1],
    hue=comb_lab,
    palette="Paired",
    alpha=0.8  # Adjust alpha to control transparency
)
plt.title("Raw kmer (k=4)")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
fig.tight_layout()
plt.savefig(kmer_4_file+".alpha_0.8_tsne.plot.png", dpi=300)

plt.show()
plt.close()
```

```python
# coverage
```

```python
coverage_file='../results/features/aug0_datacoverage_mean.tsv'
cov_df=pd.read_csv(coverage_file,sep='\t',index_col=0)


namelist = pd.read_csv(coverage_file, sep='\t', usecols=range(1)).values[:, 0]
mapObj = dict(zip(range(len(namelist)), namelist))
covMat = cov_df.values
covMat = covMat + 1e-5
covMat = covMat / covMat.max(axis=0)[None, :]


covMat_df=pd.DataFrame(covMat,index=namelist)


covMat_top10bin_df=covMat_df[covMat_df.index.isin(contigs_top10bin_list)]


covMat_top10bin_label=[]
for key in list(covMat_top10bin_df.index):
    covMat_top10bin_label.append(seq_to_binid[key])


# Visualize embeddings using tSNE
comb_emb = covMat_top10bin_df.values
comb_lab = np.array(covMat_top10bin_label)

comb_tsne = TSNE().fit_transform(comb_emb)

fig, ax = plt.subplots()

sns.scatterplot(
    x=comb_tsne[:, 0],
    y=comb_tsne[:, 1],
    hue=comb_lab,
    palette="Paired",
    alpha=0.8  # Adjust alpha to control transparency
)

plt.title("Raw coverage")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
# plt.show()
fig.tight_layout()

plt.savefig(coverage_file + ".alpha_0.8_tsne.plot.png", dpi=300)

```

```python
# coverage and kmer
```

```python
coverage_file='../results/features/aug0_datacoverage_mean.tsv'
namelist = pd.read_csv(coverage_file, sep='\t', usecols=range(1)).values[:, 0]
mapObj = dict(zip(namelist, range(len(namelist))))
cov_file=coverage_file
com_file='../results/features/kmer_4_f0.csv'

covHeader = pd.read_csv(cov_file, sep='\t', nrows=1)
shuffled_covMat = pd.read_csv(cov_file, sep='\t', usecols=range(1, covHeader.shape[1])).values
shuffled_namelist = pd.read_csv(cov_file, sep='\t', usecols=range(1)).values[:, 0]

covIdxArr = np.empty(len(mapObj), dtype=np.int)
for contigIdx in range(len(shuffled_namelist)):
    if shuffled_namelist[contigIdx].split('_aug')[0] in mapObj:
        covIdxArr[mapObj[shuffled_namelist[contigIdx].split('_aug')[0]]] = contigIdx
covMat = shuffled_covMat[covIdxArr]


compositHeader = pd.read_csv(com_file, sep=',', nrows=1)
shuffled_compositMat = pd.read_csv(com_file, sep=',', usecols=range(1, compositHeader.shape[1])).values
shuffled_namelist = pd.read_csv(com_file, sep=',', usecols=range(1)).values[:, 0]

covIdxArr = np.empty(len(mapObj), dtype=np.int)
for contigIdx in range(len(shuffled_namelist)):
    if shuffled_namelist[contigIdx].split('_aug')[0] in mapObj:
        covIdxArr[mapObj[shuffled_namelist[contigIdx].split('_aug')[0]]] = contigIdx
compositMat = shuffled_compositMat[covIdxArr]


covMat = covMat + 1e-5
covMat = covMat / covMat.max(axis=0)[None, :]

compositMat = compositMat + 1
compositMat = compositMat / compositMat.sum(axis=1)[:, None]


X_ts = np.hstack((covMat, compositMat))

combine_df=pd.DataFrame(X_ts,index=namelist)


combine_top10bin_df=combine_df[combine_df.index.isin(contigs_top10bin_list)]

combine_top10bin_label=[]
for key in list(combine_top10bin_df.index):
    combine_top10bin_label.append(seq_to_binid[key])


# Visualize embeddings using tSNE
comb_emb = combine_top10bin_df.values
comb_lab = np.array(combine_top10bin_label)


comb_tsne = TSNE().fit_transform(comb_emb)

fig, ax = plt.subplots()

sns.scatterplot(
    x=comb_tsne[:, 0],
    y=comb_tsne[:, 1],
    hue=comb_lab,
    palette="Paired",
    alpha=0.8  # Adjust alpha to control transparency
)

plt.title("Raw (kmer:k=4;coverage)")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0)
# plt.show()
fig.tight_layout()
plt.savefig("../results/features/Raw_4mer_coverage.alpha_0.8_tsne.plot.png", dpi=300)


plt.show()
plt.close()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


def plot_pathofact_res():
    subset = pd.DataFrame([[8,5,5,8],[7,6,7,12],[14,11,12,18]],
                          index=['Water Group1', 'Water Group2','Water Group3'],
                          columns=['MetaBAT2','MetaDecoder','SemiBin2','COMEBin'])


    ax = subset.plot(kind="bar",
                     legend=False, figsize=(3, 4), color = [ '#BDE6FA','#87C3EC','#5AA2DA','#446FA3'])

    # for p in ax.patches:
    #     ax.annotate(str(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom',fontsize=6)

    ax.legend(['MetaBAT2','MetaDecoder','SemiBin2','COMEBin'],
              loc='upper left', fontsize=8)
    y_label=[0,5,10,15,20,25]
    ax.set_yticks(ticks=y_label)
    ax.set_yticklabels(labels=y_label, fontsize=8, color='black')

    ax.set_xticklabels(labels=subset.index, rotation=30,
                       minor=False, fontsize=8, color='black')
    ax.set_ylabel('Number of Potential PARB', fontsize=10, color='black')

    plt.savefig('./figures/PARB_res.pdf', dpi=300, bbox_inches='tight')

```
